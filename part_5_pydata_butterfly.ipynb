```python
import seaborn as sns
from matplotlib import patches, pyplot as plt
from ipywidgets import interact

# set initial position, velocity, and radius of two circles, a and b
a_x, a_y = -2 / 3, 0
a_vx, a_vy = 1, 0
a_r = 1
b_x, b_y = 1 + 1 / 2, 1
b_vx, b_vy = 0, 0
b_r = 1


def plot_circles_given_time(clock):
    # calculate the position of the two circles at time t
    aprime_x = a_x + a_vx * clock
    aprime_y = a_y + a_vy * clock
    bprime_x = b_x + b_vx * clock
    bprime_y = b_y + b_vy * clock

    # calculate the distance between the two circles at time t
    distance = ((aprime_x - bprime_x) ** 2 + (aprime_y - bprime_y) ** 2) ** 0.5

    # print the distance between the two circles at time t
    print(f"distance between centers {distance:.3f} =?= {a_r + b_r}")
    print(f"aprime_x = {aprime_x}")

    # plot using matplotlib and Seaborn-style
    sns.set()
    figure, axes = plt.subplots()
    figure.set_dpi(100 * 4)
    figure.tight_layout()
    axes.set_title(f"clock = {clock:.1f}", loc="left")
    axes.set_aspect(1)
    axes.set_xlim(-2.0, 5.0)
    axes.set_ylim(-1.5, 2.5)
    axes.add_patch(patches.Circle((aprime_x, aprime_y), a_r, color="tab:red"))
    axes.add_patch(patches.Circle((bprime_x, bprime_y), a_r, color="tab:blue"))
    axes.annotate(
        "",
        xy=(aprime_x + a_vx, aprime_y + a_vy),
        xytext=(aprime_x, aprime_y),
        arrowprops=dict(arrowstyle="->", color="darkblue"),
    )
    axes.annotate(
        "",
        xy=(bprime_x + b_vx, bprime_y + b_vy),
        xytext=(bprime_x, bprime_y),
        arrowprops=dict(arrowstyle="->", color="darkblue"),
    )


# plot the circles at time clock = 0, 1
plot_circles_given_time(0)
plot_circles_given_time(1)

# run the interactive plot
interact(plot_circles_given_time, clock=(-1, 4, 0.1), continuous_update=False)
```
