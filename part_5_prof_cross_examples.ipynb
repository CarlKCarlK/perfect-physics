```python
import pandas as pd
from sympy import sin, cos, pi
from perfect_physics import Circle, World

angle = 60 * pi / 180
v0 = 1

ball2 = Circle(x=0,y=0,r=1,vx=0,vy=0,m=1)
ball3 = Circle(x=2,y=0,r=1,vx=0,vy=0,m=1)
ball1 = Circle(x=-cos(angle)*4,y=-sin(angle)*4,r=1,vx=v0*cos(angle),vy=v0*sin(angle),m=1)
world = World([ball1,ball2,ball3], xlim=(-6, 4), ylim=(-5, 2))

world.run_in_place(2, show=True)

# put vx and vy in a dataframe
df = pd.DataFrame(
    [(circle.vx,circle.vy) for circle in world.circle_list],
    columns=["vx", "vy"],
    index=[f"ball{index}" for index in range(1,4)],
)
df
```

```python
import pandas as pd
from sympy import sin, cos, pi
from perfect_physics import Circle, World

angle = 60 * pi / 180
v0 = 1

ball1 = Circle(x=0,y=0,r=1,vx=0,vy=0,m=1)
ball2 = Circle(x=2,y=0,r=1,vx=0,vy=0,m=1)
ball3 = Circle(x=4,y=0,r=1,vx=0,vy=0,m=1)
ball0 = Circle(x=-cos(angle)*4,y=-sin(angle)*4,r=1,vx=v0*cos(angle),vy=v0*sin(angle),m=1)
world = World([ball0,ball1,ball2,ball3], xlim=(-4, 6), ylim=(-5, 2))

world.run_in_place(3, show=True)

df = pd.DataFrame(
    [(circle.vx,circle.vy) for circle in world.circle_list],
    columns=["vx", "vy"],
    index=[f"ball{index}" for index in range(4)],
)
df
```
