```python
from sympy import Rational
from perfect_physics import Circle, World

for seed in range(2):
    left = Circle(x=-3,y=0,r=1,vx=1,vy=0,m=Rational(1,5))
    middle = Circle(x=0,y=0,r=1,vx=0,vy=0,m=1)
    right = Circle(x=3,y=0,r=1,vx=-1,vy=0,m=5)
    world = World([left,middle,right], xlim=(-10, 10), ylim=(-2, 2), rng=seed)
    world.run_in_place(2,show=True, font_scale=1)
    print(f"seed={seed},vx={[circle.vx for circle in world.circle_list]}")
    pass
```
