```python
from perfectphysics import World, Circle, Wall

big_radius = 10
world_width = 40
folder = f"m:/deldir/perfect_physics/part6/two_size{big_radius}"

big = Circle(x=world_width // 2, y=0, r=big_radius, vx=1, vy=0, m=big_radius**3)
little = Circle(x=big.x - big_radius - 1, y=0, r=1, vx=1, vy=0, m=1)
circle_list = [big, little]
wall_list = [
    Wall(x0=0, y0=0, x1=0, y1=1),
    Wall(x0=world_width, y0=0, x1=world_width, y1=1),
]
world = World(
    circle_list,
    wall_list,
    xlim=(-1, world_width + 1),
    ylim=(-big_radius - 1, big_radius + 1),
)
if True:
    world.run_in_place(2, show=True)
    print(world.circle_list)
else:
    world.run_to_file(folder, 2000)
    World.render(folder, speed_up=10)
```
