```python
# create billiards, but with jitter

import numpy as np
from perfect_physics import World
from sympy import S
import logging
logging.basicConfig(level=logging.INFO)


def jitter_world(seed = 0, resolution=100, rows=3):

    rng = np.random.default_rng(seed)

    def random_point(rng, resolution, abs=False):
        lower = -resolution if not abs else 0
        return S(rng.integers(lower, resolution,endpoint=True))/resolution

    world = World.billiards(rows=rows, rng=rng)
    for circle in world.circle_list:
        # print(circle.r)
        circle.r = circle.r - random_point(rng, resolution, abs=True)/100
        circle.r *= S(99)/100
        # print(circle.r)
        print(circle.x, circle.y)
        circle.x += random_point(rng, resolution)/100
        circle.y += random_point(rng, resolution)/100
        print(circle.x, circle.y)
    return world
jitter_world().show()
pass
```
